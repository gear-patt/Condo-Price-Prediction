```python
import pandas as pd
data = pd.read_csv('dataset.csv')
data_original = data.copy()
data.head(10)
```

## Built Year

```python
import re
int(re.findall(r'\d+', data['built_year_list'][0])[0])
```

```python
built_year_list = []
for year in data['built_year_list']:
    try:
        built_year_list.append(int(re.findall(r'\d+', year)[0]))
    except:
        built_year_list.append(None)

built_year_list[:5]
```

```python
data['built_year_list'] = built_year_list
data.head(10)
```

## Location

```python
data['location_list'][0].split(', ')
```

```python
province = []
sub_area = []
for location in data['location_list']:
    try:
        province.append(location.split(', ')[1])
        sub_area.append(location.split(', ')[0])
    except:
        sub_area.append(None)
        province.append(location)

print(province[:5])
print(sub_area[:5])
```

```python
data.insert(1, "Province", province)
data.insert(2, "Sub_Area", sub_area)
data.head()
```

```python
floor_list = []
for floor in data['floor_list']:
    if type(floor) != float:
        if floor == '12A':
            floor = '12'
        elif floor == '12a':
            floor = '12'
        elif floor == 'G':
            floor = '1'
        elif floor == '11-20':
            floor = '15'
        elif floor == 'ุุุ6':
            floor = '6'
        elif floor == '5-10':
            floor = '7'
        elif floor == '\u200b26' or floor == '1970-01-01T00:00:00.007Z' or floor == '\u200b4':
            floor_list.append(None)
            continue
        elif floor == '17TH':
            floor = '17'
        elif floor == '10B':
            floor = '10'
        elif floor == '17-18':
            floor = '17'
        elif floor == '21-50':
            floor = '35'
        elif floor == '33-34':
            floor = '33'
        floor_list.append(int(floor))
    else:
        floor_list.append(None)
        
data['floor_list'] = floor_list
data.head()
```

```python
data.head()
```

```python
data['price_list'] = [int(price[1:].replace(',','')) for price in data['price_list']]
data.head()
```

```python
data['bedroom_list'] = [int(bedroom.split(' ')[0]) for bedroom in data['bedroom_list']]
data['bathroom_list'] = [int(bathroom.split(' ')[0]) for bathroom in data['bathroom_list']]
data['area_list'] = [float(area.split(' ')[0]) for area in data['area_list']]
data.head()
```

```python
distance_to_transport_list = []
for distance in data['distance_to_transport_list']:
    try:
        distance_to_transport_list.append(int(distance.split(' ')[0]))
    except:
        distance_to_transport_list.append(None)
        
distance_to_transport_list[:5]
```

```python
data['distance_to_transport_list'] = distance_to_transport_list
data.head()
```

```python
facilities_list = []
for facilities in data['facilities_list']:
    if type(facilities) != float:
        facilities_list.append(len(facilities[1:-1].split(", ")))
    else:
        facilities_list.append(None)

facilities_list
```

```python
data['facilities_list'] = facilities_list
data.head()
```
